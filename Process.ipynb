```python
from multiprocessing import Process
import os
import time
def Square_numbers():
  for i in range(100):
    i*i 
    time.sleep(0.1)

processes=[]
num_pro=os.cpu_count()                       #number of Processes, good number is numer of CPU

#Create Processes
for i in range(num_pro):
  p=Process(target=Square_numbers)
  processes.append(p)

#start
for p in processes:
  p.start()
# Join (wait for the process)
for p in processes:
  p.join()

print("end main")


```
